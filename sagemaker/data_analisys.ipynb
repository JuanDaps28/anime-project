```python
from utils import DataDescriptor
```

```python
data_descriptor = DataDescriptor(
    data_query="""
        SELECT
            r.rating,
            c.genre,
            c.type,
            c.episodes
        FROM animes_ratings r
        INNER JOIN animes_catalog c
        ON
            r.anime_id = c.anime_id
        WHERE
            c.deleted = False
    """
)
```

```python
data_descriptor.get_histograms(
    variables=["rating", "episodes"],
    bins_amount={
        "rating": 10,
        "episodes": 50
    }
)
```

```python
data_descriptor.get_barplots(
    variables=["type"]
)
```

```python
data_descriptor.get_boxplots(
    target_variable="episodes"
)
```

```python
data_descriptor.get_boxplots(
    target_variable="rating",
    split_variables="type"
)
```

```python
data_descriptor.get_scatterplots_features_vs_target(
    target_variable="rating",
    variables=["episodes", "type"]
)
```

```python
data_descriptor.get_corr_matrix(
    variables=["episodes", "rating"]
)
```
